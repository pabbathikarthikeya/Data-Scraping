```python
import requests
import pandas as pd

def scrape_leads_with_serpapi(api_key, keyword, location, num_results=30):
    query = f'site:linkedin.com/in/ "{keyword}" "{location}"'

    params = {
        "engine": "google",
        "q": query,
        "num": num_results,
        "api_key": api_key
    }

    print(f"Searching: {query}")

    response = requests.get("https://serpapi.com/search", params=params)
    data = response.json()

    leads = []
    for result in data.get("organic_results", []):
        leads.append({
            "Name/Title": result.get("title"),
            "LinkedIn URL": result.get("link"),
            "Snippet": result.get("snippet")
        })

    return pd.DataFrame(leads)

# ðŸ”‘ Replace with your actual SerpAPI key
api_key = "e59d3441aefda6456b467ba272417d89ab5b5a7accc9f668f523321e5d8bc057"


#keyword = "Software Engineer"
#location = "India"
keyword = "Machine Learning Engineer"
location = "India"

df = scrape_leads_with_serpapi(api_key, keyword, location)
df.to_csv("leads.csv", index=False)
df.head()
```

```python
def score_lead(row):
    score = 0
    keywords = ['Google', 'AI', 'ML', 'Data', 'Senior', 'Principal', 'Lead', 'Manager', 'Director']
    for kw in keywords:
        if kw.lower() in row['Name/Title'].lower() or kw.lower() in row['Snippet'].lower():
            score += 2
    return score

df['Lead Score'] = df.apply(score_lead, axis=1)
df = df.sort_values(by='Lead Score', ascending=False)
df.to_csv("scored_leads.csv", index=False)
df.head()
```
